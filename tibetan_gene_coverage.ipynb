# Tibetan Gene Coverage Summary

This looks at coverage across each individual, and makes a histogram of coverage of each gene

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tib_genes = pd.read_table("../data/tib_gene_coverage_summary.csv", sep = ",")
```

```python
tib_genes
```

```python
tib_genes.columns.values
```

```python
genes = tib_genes.columns.values[1:]
print(genes)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
i = 1  
plt.figure(figsize=(15,9))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=.5, hspace=.5)
for gene in genes:
    plt.subplot(4,6,i)
    plt.hist(tib_genes[gene])
    plt.xlabel(gene + ' coverage')
    plt.ylabel('count')
    i+=1
plt.savefig("tibetan_gene_coverage.png", dpi=300)
```
